```python
from scraper import *
import os
import json
import urllib.request
import random
import time

import sys
sys.path.append("..")
from utils.download_utils import *
```

```python
# Load the existing JSON file at the start
filename = 'trip_advisor.json'
# Create a directory for downloads if it doesn't exist
download_path = '../downloads'
os.makedirs(download_path, exist_ok=True)

if os.path.exists(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        all_shop_data = json.load(f)
```

```python
# Categories to accept
# acceptable_categories = {'Chinese', 'Healthy', 'Asian', 'Seafood', 'Szechuan', 'Cantonese', 'Dessert', 'Quick Bites', 'International', 'Fusion', 'Bakeries', 'Fast food', 'Hong Kong'}

excluded_categories = {'Japanese', 'Singaporean', 'Indian', 'Korean', 'American', 'Type not available', 'Thai', 'Vietnamese', 'British'}

def filter_restaurants(restaurant_list):
    filtered_restaurants = []

    for restaurant in restaurant_list:
        # Split types into individual categories and strip extra white space
        categories = {category.strip() for category in restaurant['type'].split(',')}
        
        
        if categories.intersection(excluded_categories):
            a = 1
            # print(f"Excluded: {restaurant['name']} has categories {categories}")
        else:
            filtered_restaurants.append(restaurant)
            

    return filtered_restaurants

# Get the filtered list
filtered_list = filter_restaurants(all_shop_data)
```
